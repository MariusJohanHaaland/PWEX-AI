```python
#Readme:
#Example of Venn diagram


# Import packages
import pandas as pd
import numpy as np

# Dataset 1
d1 = {'Customer_id':pd.Series([1,2,3,4,5,6]),
  'Product':pd.Series(['Oven','Oven','Oven','Television','Television','Television'])}
df1 = pd.DataFrame(d1)

# Dataset 2
d2 = {'Customer_id':pd.Series([2,4,6,7,8]),
    'State':pd.Series(['California','California','Texas','New York','Indiana'])}
df2 = pd.DataFrame(d2)

#Procedure:
# how='left', indicator will either be 'both' or 'left_only'
result = df1.merge(df2, how='outer', indicator=True)

# result as list
#result[0].tolist()
```

```python
#Import libraries for Venn Diagrams
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib import pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
import numpy as np


# Simple Venn Diagram Unweighted (meaning the circles have the same size)
venn2_unweighted(subsets = (x, y, z), set_labels = ('Group A', 'Group B'), set_colors=('forestgreen', 'skyblue'), alpha = 0.5);
# dashed lines around the circles
venn2_circles(subsets = (1, 1, 1), linestyle='dashed', linewidth=2, color='k');
```
